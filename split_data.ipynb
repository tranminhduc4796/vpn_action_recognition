```
import cv2
from tqdm import tqdm
import zipfile
from imutils import paths
import shutil
import os
```

```
def extractImages(pathIn, pathOut):
  cap= cv2.VideoCapture(pathIn)
  i=0
  img_name = pathIn.split('/')[3].split('_')[0]
  while cap.isOpened():
      ret, frame = cap.read()
      if ret == False:
          break
      cv2.imwrite(pathOut + '{0}_frame_{1}.jpg'.format(img_name, i), frame)
      i+=1
```

```
for i in range(14, 18):

  #zip file to folder
  zip_ref = zipfile.ZipFile("/content/gdrive/MyDrive/NTU_RGB_video/nturgbd_rgb_s{}.zip".format(str(i).zfill(3)), 'r')
  zip_ref.extractall("/tmp")
  zip_ref.close()

  #split video to images
  video_path = list(paths.list_files('/tmp/nturgb+d_rgb'))
  for i in tqdm(video_path):
    label = i.split('A')[-1].split('_')[0]
    save_path = '/content/gdrive/MyDrive/NTU_RGB_video/A{}/'.format(label)
    if label in ['008', '009', '027', '058', '059', '060']:
      extractImages(i, save_path)
      
  #remove folder    
  shutil.rmtree('/tmp/nturgb+d_rgb')
```
